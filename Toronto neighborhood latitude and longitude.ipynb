```python
import pandas as pd # library for data analsysis
import googlemaps
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
```

```python
csv_file_url='https://raw.githubusercontent.com/fabioantonini/Coursera_Capstone/master/toronto_postal_codes.csv'
df_postcodes = pd.read_csv(csv_file_url, error_bad_lines=False)
```

```python
df_postcodes.head(20)
#df_postcodes.shape
```

Iterate through the Toronto's Postal Codes

```python
csv_latlong_url='https://cocl.us/Geospatial_data'
df_latlong = pd.read_csv(csv_latlong_url, error_bad_lines=False)

# instantiate the dataframe
column_names = ['PostCode', 'Borough', 'Neighborhood', 'Latitude', 'Longitude'] 
df_postcode_latlong = pd.DataFrame(columns=column_names)

for index, row in df_postcodes.iterrows():
    # extract Postcode, Borough, Neighbourhood
    postcode = row['Postcode']
    borough = row['Borough']
    neighbourhood_name = row['Neighbourhood']
    # retrieve latitude and longitude from the df_latlong dataframe read from the csv file
    neighbourhood_lat = df_latlong.loc[row['Postcode'] == df_latlong['Postal Code'],:]['Latitude'].values[0]
    neighbourhood_lon = df_latlong.loc[row['Postcode'] == df_latlong['Postal Code'],:]['Longitude'].values[0]
    # append all the data to the new dataframe including Postcode, Borough, Neighbourhood, Latitude and Longitude
    df_postcode_latlong = df_postcode_latlong.append({'PostCode': postcode, 
                                'Borough': borough,
                                'Neighborhood': neighbourhood_name,
                                'Latitude': neighbourhood_lat,
                                'Longitude': neighbourhood_lon}, ignore_index=True)
df_postcode_latlong.head(12)
```

```python
df_postcode_latlong.to_csv(r'toronto_postcodes_latlon.csv', index=False)
```
